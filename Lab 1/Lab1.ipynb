```python
from __future__ import division
from posixpath import split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Excercise 1 function
def scale_data(data_array):
  stdev = np.std(data_array)
  output_array = []
  mean = np.mean(data_array)
  for i in range(len(data_array)):
    output_array.append((data_array[i] - mean)/stdev)
  return output_array


# Excercise 2 function
def split_data(arr, split_proportions, axis):
  output_array = ([])
  errcatch = 0

  # Normalize proportions
  for i in range(len(split_proportions)):
    errcatch += split_proportions[i]
  if(errcatch > 1):
    for i in range(len(split_proportions)):
      split_proportions[i] = split_proportions[i]/errcatch

  # Find division sizes
  division_sizes = []
  truesplit = 0
  for split in split_proportions[:-1]:
    truesplit += int(np.shape(arr)[axis] * split)
    division_sizes.append(truesplit)

  if axis == 1:
    output_array = np.hsplit(arr, division_sizes)
  else:
    output_array = np.vsplit(arr, division_sizes)
  return output_array

```

```python
# Excercise 1 function test with x

hgcal = pd.read_csv('hgcal.csv')

scaled_hgcal = []
scaled_hgcal = scale_data(hgcal['x'])

plt.hist(scaled_hgcal)
plt.show()
```
